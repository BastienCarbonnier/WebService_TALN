```python
import spacy
import treetaggerwrapper

def posTaggingSpacy(text) :
    nlp = spacy.load("fr_core_news_sm")
    doc = nlp(text)

    list_t =[]

    for token in doc:
        t = {"mot":token.text, "nature": tags[token.pos_],"tag":token.pos_}
        list_t.append(t)
    return list_t



tagger = treetaggerwrapper.TreeTagger(TAGLANG='fr',TAGOPT="-prob -threshold 0.7 -token -lemma -sgml -quiet")

def tagToObj(sentArray): 
    sentTab = []
    for t in sentArray: 
        if len(t) >1 :
            sentTab.append({"mot" : t[0], "nature" : tags[t[1]],"tag":t[1],"acc":t[3]})
    return sentTab

def posTaggingTreeTagger(text) :
    tags = tagger.tag_text(text)
    taggedText = treetaggerwrapper.make_tags(tags,allow_extra = True)
    return tagToObj(taggedText)
```
